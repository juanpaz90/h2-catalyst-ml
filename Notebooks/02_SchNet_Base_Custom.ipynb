## **Result analysis**

Why wasn't the Multihead attention improvement huge?

- Everything relies on the SchNet architecture. SchNet is *Distance-Only*, that means that SchNet only knows the distance between atoms. It does not know bond angles or directional geometry.

- Multihead Attention allowed the model to look at the "global picture" of the molecule. However, because the underlying SchNet embeddings still lack 3D angular awareness, the MHA layer is trying to extract global context from fundamentally limited local features.

In the IS2RE task, I'm basically trying to predict the final relaxed energy without taking in consideration all the physics from the elements. SchNet does not use angles, shapes or anything else aside of distances thats why results are not so great even with the use of MHA.
