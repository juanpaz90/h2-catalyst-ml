**About Optimizer and criterion:**
- The `criterion` (`L1Loss`) defines what the goal is (minimize absolute error), which is dictated by the laws of the specific physics problem I am solving. Therefore, it shouldn't be changed.
- The `optimizer` (`Adam` vs. `AdamW`), learning rate, and hidden channels define how the model navigates the math to achieve that goal. Therefore, they are perfectly suited for Optuna to tune
